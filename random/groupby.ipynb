```python
import pandas as pd

# Create a sample DataFrame
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B', 'D'],
        'Appt_Date': [['13 Feb 2013', 20, 10, '19 May 2007', 20, 40, '19 May 2007', '13 Feb 2013', 0],
        'Rank': ['1LT', 'CPT', '1LT', 'MAJ', 'CPT', '1LT', 'MAJ', 'CPT', 'LTC']}
df = pd.DataFrame(data)
display(df)
# Group by 'Category' and get the mode of 'Appt_Date'
mode_result = df.groupby('Category')['Appt_Date'].agg(lambda x: x.mode()[0])

print(mode_result)
```

```python
import pandas as pd

data = {'Name':['Bob','Dave','Bob','Charlie','Dave','Bob','Charlie','Dave','Eve','Bob'],
        'Snapshot_Date':['31 Dec 2015','31 Dec 2015','31 Mar 2018',
                         '31 Dec 2015','31 Mar 2018','29 Sep 2021','31 Mar 2018',
                         '29 Sep 2021','31 Dec 2015','31 Jun 2023'],
        'Appt_Date':['13 Feb 2013','10 Oct 2009','13 Feb 2013','19 May 2007','10 Oct 2009','20 Dec 2013','19 May 2007','13 Feb 2013','12 Aug 2002','20 Dec 2013'],
        'Rank':['1LT','CPT','1LT','MAJ','CPT','1LT','MAJ','CPT','LTC','1LT']}

df = pd.DataFrame(data)
df['Appt_Date'] = pd.to_datetime(df['Appt_Date'])
df.groupby('Name')[['Appt_Date']].agg(lambda x: x.mode()[0])
df = df[['Snapshot_Date','Name','Appt_Date','Rank']]
desired_order = ['1LT','CPT','MAJ','LTC','COL']
# Using pd.Categorical()
df['Rank'] = pd.Categorical(df['Rank'], categories=desired_order, ordered=True)
df.sort_values(by=['Rank','Appt_Date'],ascending=True)
df['YG'] = df.groupby('Name')['Appt_Date'].transform(lambda x: x.mode()[0].year)
df = df.sort_values(by=['Name','Appt_Date'],ascending=True)
df
```
